```python
import numpy as np
from keras import Sequential
from keras.src.layers import Dense, SimpleRNN, LSTM, GRU, GlobalMaxPool1D
import plotly.graph_objects as go
from keras.src.optimizers import Adam
```

```python
T = 10
D = 1
X = []
Y = []


def get_label(x, i1, i2, i3):
    # x = sequence
    if x[i1] < 0 and x[i2] < 0 and x[i3] < 0:
        return 1
    if x[i1] < 0 and x[i2] > 0 and x[i3] > 0:
        return 1
    if x[i1] > 0 and x[i2] < 0 and x[i3] > 0:
        return 1
    if x[i1] > 0 and x[i2] > 0 and x[i3] < 0:
        return 1

    return 0


for t in range(5000):
    x = np.random.randn(T)
    X.append(x)
    y = get_label(x, -1, -2, -3)  # short distance
    # y = get_label(x, 0, 1, 2)  # long distance
    Y.append(y)

X = np.array(X)
Y = np.array(Y)
N = len(X)
```

```python
# Linear model
model = Sequential()
model.add(Dense(units=1, activation='sigmoid', input_shape=(T,)))

model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

r = model.fit(X, Y, epochs=100, validation_split=0.5)
```

```python
def draw_history(r):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=r.epoch, y=r.history['accuracy'], name='accuracy'))
    fig.add_trace(go.Scatter(x=r.epoch, y=r.history['val_accuracy'], name='val_accuracy'))
    fig.show()

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=r.epoch, y=r.history['loss'], name='loss'))
    fig.add_trace(go.Scatter(x=r.epoch, y=r.history['val_loss'], name='val_loss'))
    fig.show()


draw_history(r)
```

```python
# Now RNN
inputs = np.expand_dims(X, -1)

model = Sequential()
model.add(SimpleRNN(units=5))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(metrics=['accuracy'],
              optimizer=Adam(learning_rate=0.01),
              loss='binary_crossentropy')
r = model.fit(inputs, Y, epochs=100, validation_split=0.5)

draw_history(r)
```

# Conclusions
- **Linear classifier** - It is unable to solve the problem
- **SimpleRNN classifier** - Solve only short distance problem (y_value depends on latest values)
- **LSTM classifier** - Solve long distance problem

SimpleRNN have problems with long short problems because of vanishing gradient.

# Conclusions

Data given to RNN layers must have dimension
**-1 x T x D**

**-1**: 3D "blocks"- Each object for such a block on the property
**T**: Value o attributes in different time - rows in blocks
**D**: Attributes of object- columns in blocks
